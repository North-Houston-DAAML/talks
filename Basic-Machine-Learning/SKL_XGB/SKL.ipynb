one hot encoding example

---

```python
import sklearn.preprocessing
import numpy

# Creates a one hot encoding engine.

ohe = sklearn.preprocessing.OneHotEncoder()

# Creates a Numpy array with one column of car manufactures.

array = ["Toyota", "Ford", "Tesla", "Toyota", "Tesla", "Honda"]
array = numpy.array([array]).T
print("\n\nNumPy array:\n")
print(array)

# Does a one hot encoding.

print("\n\none hot encoding:\n")
print(ohe.fit_transform(array).toarray())

# Creates a NumPy array with columns of fruits and car manufacturers.

array = [["apple",  "Toyota"], 
         ["pear",   "Ford"], 
         ["banana", "Tesla"], 
         ["pear",   "Toyota"], 
         ["apple",  "Tesla"], 
         ["banana", "Honda"]]
array = numpy.array(array)
print("\n\nNumPy array:\n")
print(array)

# Does a one hot encoding.

print("\n\none hot encoding:\n")
print(ohe.fit_transform(array).toarray())
```

Creates a classification model to detect breast cancer.

Dataset contains data for several women.

---

```python
import sklearn.neighbors
import sklearn.datasets
import sklearn.model_selection
import pandas
import numpy
import matplotlib

SPLITTER = sklearn.model_selection.train_test_split

# Prepares data.

data = sklearn.datasets.load_breast_cancer()
train_inp, test_inp, train_out, test_out = SPLITTER(data["data"],
                                                    data["target"],
                                                    test_size    = 0.2,
                                                    random_state = 1234)

# Presents data summary statistics.

def describe(data):
        """
        Presents summary statistics for data.
        """

        print(f"\n\nnumber of features: {data.shape[1] - 1}")
        print(f"\n\nnumber of input output pairs: {data.shape[0]}")
        print("\n\nsome input output pairs:\n")  
        display(data.head())
        print("\n\noutput counts:\n")
        print(data.iloc[:, -1].value_counts())
        print("\n\ninput summary statistics:\n")
        display(data.iloc[:, :-1].describe()) 

data_         = [pandas.DataFrame(data["data"]), 
                 pandas.DataFrame(data["target"])]
data_         = pandas.concat(data_, axis = 1)
data_.columns = list(data["feature_names"]) + ["cancer"]
describe(data_)
```

```python
# Creates models and displays their accuracies.

train_acc = []
test_acc  = []
ks        = range(1, 100)
for k in ks:
        model = sklearn.neighbors.KNeighborsClassifier(n_neighbors = k)
        model.fit(train_inp, train_out)
        train_acc.append(model.score(train_inp, train_out))
        test_acc.append(model.score(test_inp, test_out))
matplotlib.pyplot.rcParams["figure.figsize"] = (15, 8)
matplotlib.pyplot.rcParams.update({"font.size" : 22})
matplotlib.pyplot.plot(ks, train_acc, label = "training data")
matplotlib.pyplot.plot(ks, test_acc,  label = "testing data")
matplotlib.pyplot.xlabel("k")
matplotlib.pyplot.ylabel("accuracy")
matplotlib.pyplot.legend()
matplotlib.pyplot.show()
```

---
same but with a decision tree model

---

```python
import sklearn.tree

PRF1 = sklearn.metrics.precision_recall_fscore_support

# Creates models.

model = sklearn.tree.DecisionTreeClassifier(max_depth = 4,
                                            random_state = 1234)
model.fit(train_inp, train_out)

# Displays accuracy reports.

def report_class(model, test_inp, test_out):
        preds = model.predict(test_inp)
        acc   = 100 * sklearn.metrics.accuracy_score(test_out, preds)
        prf1  = PRF1(test_out, preds)
        print(f"\n\naccuracy: {acc}")
        print(f"\n\nprecision: {prf1[0]}")
        print(f"\n\nrecall: {prf1[0]}")
        print(f"\n\nF1 score: {prf1[0]}")
        print("\n\nconfusion matrix:\n")
        print(sklearn.metrics.confusion_matrix(test_out, preds))

report_class(model, test_inp, test_out)
```

---
same but with a random forest model

---

```python
import sklearn.ensemble

# Creates models.

model = sklearn.ensemble.RandomForestClassifier(n_estimators = 100,
                                                random_state = 1234)
model.fit(train_inp, train_out)

# Displays accuracy reports.

report_class(model, test_inp, test_out)
```

---
same but with a gradient boosted model

---

```python
import sklearn.ensemble

# Creates models.

model = sklearn.ensemble.GradientBoostingClassifier(random_state = 1234)
model.fit(train_inp, train_out)

# Displays accuracy reports.

report_class(model, test_inp, test_out)
```

Creates a regression model to predict RAM prices.

Dataset contains RAM prices for several years.

---

```python
import sklearn.linear_model
import sklearn.tree
import numpy
import pandas
import matplotlib

DATA_URL = "https://raw.githubusercontent.com/North-Houston-DAAML/talks/main/Basic-Machine-Learning/SKL_XGB/historical_RAM_prices_dataset"

# Prepares data.

data      = pandas.read_csv(DATA_URL, sep    = ",", dtype  = "float")
train_inp = numpy.array([data[data.date < 2000].date]).T
train_out = numpy.log([data[data.date < 2000].price]).T
test_inp  = numpy.array([data[data.date >= 2000].date]).T
test_out  = numpy.log([data[data.date >= 2000].price]).T

# Displays plots of data.

print("\n\nplot of the data:\n")

matplotlib.pyplot.rcParams["figure.figsize"] = (15, 8)
matplotlib.pyplot.rcParams.update({"font.size" : 22})
matplotlib.pyplot.xlabel("year")
matplotlib.pyplot.ylabel("log(price)")
matplotlib.pyplot.plot(train_inp, train_out)
matplotlib.pyplot.plot(test_inp, test_out)
matplotlib.pyplot.show()
```

```python
# Creates models.

model = sklearn.linear_model.LinearRegression()
model.fit(train_inp, train_out)

# Displays accuracy reports.

def report_reg(model, test_inp, test_out):
        preds = model.predict(test_inp)
        rmse  = sklearn.metrics.mean_squared_error(test_out, preds)
        mae   = sklearn.metrics.mean_absolute_error(test_out, preds)
        r2    = sklearn.metrics.r2_score(test_out, preds)
        print(f"\n\nrmse: {rmse}")
        print(f"\n\nmae: {mae}")
        print(f"\n\nR2: {mae}")

report_reg(model, test_inp, test_out)

# Displays plots of the predictions.

print("\n\nplot of the predictions:\n")

matplotlib.pyplot.rcParams["figure.figsize"] = (15, 8)
matplotlib.pyplot.rcParams.update({"font.size" : 22})
matplotlib.pyplot.plot(train_inp, train_out)
matplotlib.pyplot.plot(test_inp,  test_out)
matplotlib.pyplot.plot(train_inp, model.predict(train_inp))
matplotlib.pyplot.plot(test_inp,  model.predict(test_inp))
matplotlib.pyplot.show()
```

---
same but with a decision tree regression model

---

```python
import sklearn.tree

# Creates models.

model = sklearn.tree.DecisionTreeRegressor()
model.fit(train_inp, train_out)

# Displays accuracy reports.

report_reg(model, test_inp, test_out)

# Displays plots of the predictions.

print("\n\nplot of the predictions:\n")

matplotlib.pyplot.rcParams["figure.figsize"] = (15, 8)
matplotlib.pyplot.rcParams.update({"font.size" : 22})
matplotlib.pyplot.plot(train_inp, train_out)
matplotlib.pyplot.plot(test_inp,  test_out)
matplotlib.pyplot.plot(train_inp, model.predict(train_inp))
matplotlib.pyplot.plot(test_inp,  model.predict(test_inp))
matplotlib.pyplot.show()
```
